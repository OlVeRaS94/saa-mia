```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
grado_estudio = ctrl.Antecedent(np.arange(0, 10), 'grado de estudio')
```

```python
grado_uso_IA = ctrl.Antecedent(np.arange(0, 10), 'grado de uso de IA')
```

```python
nota = ctrl.Antecedent(np.arange(0, 10), 'nota')
```

```python
grado_estudio['bajo'] = fuzz.trimf(grado_estudio.universe, [0, 1, 3])
grado_estudio['medio'] = fuzz.trimf(grado_estudio.universe, [4, 6, 7])
grado_estudio['alto'] = fuzz.trimf(grado_estudio.universe, [8, 9, 10])
```

```python
grado_uso_IA['malo'] = fuzz.trimf(grado_uso_IA.universe, [0, 2, 3])
grado_uso_IA['regular'] = fuzz.trimf(grado_uso_IA.universe, [4, 5, 7])
grado_uso_IA['bueno'] = fuzz.trimf(grado_uso_IA.universe, [7, 9, 10])
```

```python
nota['baja'] = fuzz.trimf(nota.universe, [0, 2, 4])
nota['media'] = fuzz.trimf(nota.universe, [5, 6, 7])
nota['alta'] = fuzz.trimf(nota.universe, [8, 9, 10])
```

```python
grado_estudio.view()
grado_uso_IA.view()
nota.view()
```

```python
regla1 = ctrl.Rule(grado_estudio['bajo'], grado_uso_IA['malo'], nota['baja'])
regla2 = ctrl.Rule(grado_estudio['medio'], grado_uso_IA['regular'], nota['media'])
regla3 = ctrl.Rule(grado_estudio['alto'], grado_uso_IA['bueno'], nota['alta'])
```

```python
sistema_comportamiento = ctrl.ControlSystem([regla1, regla2, regla3])
simulador = ctrl.ControlSystemSimulation(sistema_comportamiento)
```

```python
simulador.input['nota'] = 7
```
