"Si el nombre de positius de l'alumne és elevat, el seu comportament serà 'bueno' i la seva nota serà 'alta'. Si el nombre de positius és moderat, el seu comportament serà 'regular' i la seva nota serà 'media'. Si el nombre de positius és baix, el seu comportament serà 'malo' i la seva nota serà 'baja'."
Implementació amb scikit-fuzzy:

Per exemple, si un alumne té un nombre de positius de 7 sobre 20, podem determinar a quin comportament pertany (malo, regular o bueno) i assignar una nota en conseqüència.

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

numero de positivos

```python
numero_positivos = ctrl.Antecedent(np.arange(0, 21, 1), 'numero_positivos')
```

comportamiento

```python
comportamiento = ctrl.Consequent(np.arange(0, 11, 1), 'comportamiento')
```

nota

```python
nota = ctrl.Consequent(np.arange(0,11,1), 'nota')
```

definimos las funciones de las variables

```python
numero_positivos['bajo'] = fuzz.trimf(numero_positivos.universe, [0, 0, 10])
numero_positivos['moderado'] = fuzz.trimf(numero_positivos.universe, [5, 10, 15])
numero_positivos['alto'] = fuzz.trimf(numero_positivos.universe, [10, 20, 20])
```

```python
comportamiento['malo'] = fuzz.trimf(comportamiento.universe, [0, 0, 5])
comportamiento['regular'] = fuzz.trimf(comportamiento.universe, [2, 5, 8])
comportamiento['bueno'] = fuzz.trimf(comportamiento.universe, [5, 10, 10])
```

```python
nota['baja'] = fuzz.trimf(nota.universe, [0, 0, 5])
nota['media'] = fuzz.trimf(nota.universe, [3, 5, 8])
nota['alta'] = fuzz.trimf(nota.universe, [5, 10, 10])
```

visualizar las variables

```python
numero_positivos.view()
comportamiento.view()
nota.view()
```

definir las reglas del sistema difuso

```python
regla1 = ctrl.Rule(numero_positivos['bajo'], comportamiento['malo'], nota['baja'])
regla2 = ctrl.Rule(numero_positivos['moderado'], comportamiento['regular'], nota['media'])
regla3 = ctrl.Rule(numero_positivos['alto'], comportamiento['bueno'], nota['alta'])
```

creamos el sitema del control difuso

```python
sistema_comportamiento = ctrl.ControlSystem([regla1, regla2, regla3])
simulador = ctrl.ControlSystemSimulation(sistema_comportamiento)
```

simulamos un alumno con 7 positivos

```python
simulador.input['numero_positivos'] = 7
```

simulacion

```python
simulador.compute()
```

acceder a los resultados de la simulacion

```python
comportamiento_resultado = simulador.output['comportamiento']
nota_resultado = simulador.output['nota']
```

muestra los rsultados

```python
print(f"Comportamiento: {comportamiento_resultado:.2f}")
print(f"Nota: {nota_resultado:.2f}")
```

visualiza los resultados

```python
comportamiento.view(sim=simulador)
nota.view(sim=simulador)
```
