```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
grado_estudio = ctrl.Antecedent(np.arange(0, 10), 'grado de estudio')
```

```python
grado_uso_IA = ctrl.Consequent(np.arange(0, 10), 'grado de uso de IA')
```

```python
nota = ctrl.Consequent(np.arange(0, 10), 'nota')
```

```python
grado_estudio['bajo'] = fuzz.trimf(grado_estudio.universe, [0, 2, 3])
grado_estudio['medio'] = fuzz.trimf(grado_estudio.universe, [3, 5, 7])
grado_estudio['alto'] = fuzz.trimf(grado_estudio.universe, [7, 9, 11])
```

```python
grado_uso_IA['malo'] = fuzz.trimf(grado_uso_IA.universe, [7, 9, 11])
grado_uso_IA['regular'] = fuzz.trimf(grado_uso_IA.universe, [3, 5, 7])
grado_uso_IA['bueno'] = fuzz.trimf(grado_uso_IA.universe, [0, 3, 3])
```

```python
nota['baja'] = fuzz.trimf(nota.universe, [0, 3, 4])
nota['media'] = fuzz.trimf(nota.universe, [4, 6, 7])
nota['alta'] = fuzz.trimf(nota.universe, [7, 9, 11])
```

```python
grado_estudio.view()
grado_uso_IA.view()
nota.view()
```

```python
regla1 = ctrl.Rule(grado_estudio['bajo'], grado_uso_IA['malo'], nota['baja'])
regla2 = ctrl.Rule(grado_estudio['medio'], grado_uso_IA['regular'], nota['media'])
regla3 = ctrl.Rule(grado_estudio['alto'], grado_uso_IA['bueno'], nota['alta'])
```

```python
sistema_comportamiento = ctrl.ControlSystem([regla1, regla2, regla3])
simulador = ctrl.ControlSystemSimulation(sistema_comportamiento)
```

```python
simulador.input['grado_estudio'] = 7
```

```python
simulador.compute()
```

```python
grado_IA_resultado = simulador.output['grado_uso_IA']
nota_resultado = simulador.output['nota']
```

```python
print(f"Grado_uso_IA: {grado_uso_IA_resultado:.2f}")
print(f"Nota: {nota_resultado:.2f}")
```

```python
grado_uso_IA.view(sim=simulador)
nota.view(sim=simulador)
```
