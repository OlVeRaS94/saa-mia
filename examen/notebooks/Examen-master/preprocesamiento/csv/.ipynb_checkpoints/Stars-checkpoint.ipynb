```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
file_path = 'Stars.csv'  # Cambia esta ruta si es necesario
stars_df = pd.read_csv(file_path)

```

```python
print("Información del dataset:")
print(stars_df.info())
```

```python
print("\nPrimeras filas del dataset:")
print(stars_df.head())
```

```python
# Step 1: Handling missing values
# Check missing values by column
missing_values = stars_df.isnull().sum()
```

```python
# Limpieza de datos: Manejo de valores nulos
stars_df['Temperature'] = stars_df['Temperature'].fillna(stars_df['Temperature'].median())
stars_df['L'] = stars_df['L'].fillna(stars_df['L'].median())
stars_df['R'] = stars_df['R'].fillna(stars_df['R'].median())
stars_df['A_M'] = stars_df['A_M'].fillna(stars_df['A_M'].median())
stars_df['Type'] = stars_df['Type'].fillna(stars_df['Type'].mode()[0])
stars_df['Color'] = stars_df['Color'].fillna(stars_df['Color'].mode()[0])

# Verificar que no queden valores nulos
print("\nValores nulos restantes por columna:")
print(stars_df.isnull().sum())
```

```python
# Estadísticas descriptivas
print("\nEstadísticas descriptivas del dataset:")
print(stars_df.describe(include='all'))
```

```python
# Visualización de distribuciones
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(8, 6))
```

```python
# Distribución de Temperature
sns.histplot(stars_df['Temperature'], kde=True, bins=30, ax=ax, color='blue')
ax.set_title('Temperature Distribution')
ax.set_xlabel('Temperature (K)')

```

```python
# Distribución de Luminosity (L) (escala logarítmica por su rango)
sns.histplot(stars_df['L'], kde=True, bins=30, ax=ax, color='orange', log_scale=True)
ax.set_title('Luminosity Distribution (Log Scale)')
ax.set_xlabel('Luminosity (L)')
```

```python
# Step 3: Visualizing distributions
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(18, 10))

# Temperature distribution
sns.histplot(stars_df['Temperature'], kde=True, bins=40, ax=ax, color='blue', stat='density', linewidth=2)
ax.set_title('Temperature Distribution')
ax.set_xlabel('Temperature (K)')
ax.set_ylabel('Density', fontsize=16)

ax.set_xlim(left=stars_df['Temperature'].min() - 10, right=stars_df['Temperature'].max() + 10)
ax.set_ylim(bottom=0)

ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.2f}'))

ax.tick_params(axis='both', which='major', labelsize=14)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
# Distribución de Radius (R)
sns.histplot(stars_df['R'], kde=True, bins=30, ax=ax, color='green')
ax.set_title('Radius Distribution')
ax.set_xlabel('Radius (R)')
```

```python
# Distribución de Absolute Magnitude (A_M)
sns.histplot(stars_df['A_M'], kde=True, bins=30, ax=ax, color='purple')
ax.set_title('Absolute Magnitude Distribution')
ax.set_xlabel('Absolute Magnitude (A_M)')
```

```python
# Distribución de Type
sns.countplot(x='Type', data=stars_df, ax=ax, hue='Type', dodge=False)
ax.set_title('Star Types Distribution')
ax.set_xlabel('Type')
ax.legend_.remove()  # Si no deseas mostrar la leyenda
```

```python
# Distribución de Spectral_Class
sns.countplot(x='Spectral_Class', data=stars_df, ax=ax, hue='Spectral_Class', dodge=False)
ax.set_title('Spectral Class Distribution')
ax.set_xlabel('Spectral Class')
```

```python
plt.tight_layout()
plt.show()
```
