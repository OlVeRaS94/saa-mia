# Enunciat: Predicció de la Temperatura en Funció de l'Altura

Vols modelar la **relació entre l'altura d'una ciutat sobre el nivell del mar** i la **temperatura mitjana mensual**. Utilitzant un model de **regressió lineal**, la variable independent serà l'**altitud** (en metres) i la variable dependent serà la **temperatura mitjana mensual** (en °C). La relació esperada és que a major altura, menor serà la temperatura. Es vol crear un model de regressió lineal amb dades simulades, afegint una mica de soroll aleatori.

## Dades simulades:

| Altura (m) | Temperatura mitjana mensual (°C) |
|-------------|----------------------------------|
| 0           | 20.5                             |
| 200         | 18.9                             |
| 400         | 17.4                             |
| 600         | 15.8                             |
| 800         | 14.2                             |
| 1000        | 12.6                             |
| 1200        | 11.0                             |
| 1400        | 9.5                              |
| 1600        | 8.0                              |
| 1800        | 6.5                              |

## Tasques:

1. Genera un conjunt de dades amb **altura** (en el rang de 0 a 2000 metres).
2. Genera la variable dependent **temperatura** seguint la fórmula de la regressió lineal: `temperatura = 20 - 0.006 * altura`, afegint una mica de soroll aleatori per fer-ho més realista.
3. Utilitza `sklearn` per crear un model de regressió lineal i ajustar-lo a les dades simulades.
4. **Representa gràficament les dades** i **la funció de regressió lineal ajustada**.
5. Calcula l'**error quadràtic mitjà (MSE)** i el **coeficient de determinació \( R^2 \)** per avaluar la qualitat del model.
6. Prediu la temperatura mitjana per a una ciutat situada a 1500 metres d'altura.


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Generació de dades simulades
np.random.seed(42)
alturas = np.random.randint(0, 2001, size=(500, 1))  # 20 mostres d'altura entre 0 i 2000 metres
temperatures = 20 - 0.006 * alturas + np.random.normal(0, 1, size=(500, 1))  # Afegint soroll aleatori
```

```python
model = LinearRegression()
model.fit(alturas, temperatures)
```

```python
temperatures_pred = model.predict(alturas)
```

```python
mse = mean_squared_error(temperatures, temperatures_pred)
r2 = r2_score(temperatures_pred,temperatures)
```

```python
print(f"Error quadratic Mitja (MSE): {mse:.2f}")
print(f"coeficient de determinacio (r²):{r2:.2f}")
```

```python
plt.scatter(alturas, temperatures, color='blue', label='Dades simulades', alpha=0.5)
plt.plot(alturas, temperatures_pred, color='red', label='Model de regressio')
plt.title('Relacio entre altura i temperatura')
plt.xlabel('Altura(m)')
plt.ylabel('Temperatura(°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
altura_1500 = np.array([[1500]])
prediccio_1500 = model.predict(altura_1500)
print(f"La temperatura mitjana en una ciutat a 1500 m de altura es de {prediccio_1500[0][0]:.2f}ºC")
```
