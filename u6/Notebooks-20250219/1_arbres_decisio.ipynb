# Arbres de decisió. Exemples.

Anem a crear un arbre de decisió per al dataset Iris, que vam treballar a la unitat anterior.

```python
# Importem les llibreries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris(as_frame=True)
X_iris = iris.data[["petal length (cm)", "petal width (cm)"]].values
y_iris = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X_iris, y_iris)
```

Per tal de visualitzar l'arbre de decisió fem ús de Graphviz, un paquet de visualització de gràfics de codi obert, que treballa amb el format .dot, però que es pot convertir a altres formats.

```python
# Instal·lem primer graphviz, que ens servirà per dibuixar els arbres
# pip install graphviz
```

I ara dibuixem l'arbre de decisió resultant amb:

```python
from sklearn.tree import export_graphviz
from graphviz import Source
from IPython.display import display

def dibuixaArbre(tree_clf, feature_names, class_names=None):

    imatge=export_graphviz(
            tree_clf,
            #feature_names=["petal length (cm)", "petal width (cm)"],
            feature_names=feature_names,
            class_names=class_names,
            #class_names=iris.target_names,
            rounded=True,
            filled=True
        )

    graph = Source(imatge)
    display(graph)
```

```python
dibuixaArbre(tree_clf, ["petal length (cm)", "petal width (cm)"], iris.target_names)
```

```python
# Representació gràfica dels resultats

import numpy as np
import matplotlib.pyplot as plt

# Informació per al format
from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#fafab0', '#9898ff', '#a0faa0'])
plt.figure(figsize=(8, 4))

lengths, widths = np.meshgrid(np.linspace(0, 7.2, 100), np.linspace(0, 3, 100))
X_iris_all = np.c_[lengths.ravel(), widths.ravel()]
y_pred = tree_clf.predict(X_iris_all).reshape(lengths.shape)
plt.contourf(lengths, widths, y_pred, alpha=0.3, cmap=custom_cmap)
for idx, (name, style) in enumerate(zip(iris.target_names, ("yo", "bs", "g^"))):
    plt.plot(X_iris[:, 0][y_iris == idx], X_iris[:, 1][y_iris == idx],
             style, label=f"Iris {name}")


tree_clf_deeper = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_clf_deeper.fit(X_iris, y_iris)
th0, th1, th2a, th2b = tree_clf_deeper.tree_.threshold[[0, 2, 3, 6]]
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.plot([th0, th0], [0, 3], "k-", linewidth=2)
plt.plot([th0, 7.2], [th1, th1], "k--", linewidth=2)
plt.plot([th2a, th2a], [0, th1], "k:", linewidth=2)
plt.plot([th2b, th2b], [th1, 3], "k:", linewidth=2)
plt.text(th0 - 0.05, 1.0, "Profunditat=0", horizontalalignment="right", fontsize=15)
plt.text(3.2, th1 + 0.02, "Profunditat=1", verticalalignment="bottom", fontsize=13)
plt.text(th2a + 0.05, 0.5, "(Profunditat=2)", fontsize=11)
plt.axis([0, 7.2, 0, 3])
plt.legend()

plt.show()
```

```python
# Per fer prediccions:
categoria=tree_clf.predict([[5, 1.5], [2, 1]])

print(categoria)

# Ens torna un Array amb la categoria, per obtindre el nom de la classe
print(iris.target_names[categoria])
```

```python
# Per obtenir la probabilitat de cada classe
tree_clf.predict_proba([[5, 1.5], [2, 1]]).round(3)
```

## Regularitzant models

```python
# Importem la funció make_moons, per generar mitges llunes
from sklearn.datasets import make_moons

# Creem un conjunt d'entrenament amb 150 mostres
X_moons, y_moons = make_moons(n_samples=150, noise=0.2, random_state=42)

# Creem dos classificadors:

# Un sense regularitzar
tree_clf1 = DecisionTreeClassifier(random_state=42)

# I un regularitzat, ajustant min_samples_leaf
tree_clf2 = DecisionTreeClassifier(min_samples_leaf=5, random_state=42)

# Entrenem ambdós models amb els conjunts d'entrenament
tree_clf1.fit(X_moons, y_moons)
tree_clf2.fit(X_moons, y_moons)
```

```python
# Veiem l'arbre resultant sense regularitzar
dibuixaArbre(tree_clf1, ["petal length (cm)", "petal width (cm)"], iris.target_names)
```

```python
# I el regularitzat
dibuixaArbre(tree_clf2, ["petal length (cm)", "petal width (cm)"], iris.target_names)
```

```python
# Veiem els límits de decisió

def plot_decision_boundary(clf, X, y, axes, cmap):
    x1, x2 = np.meshgrid(np.linspace(axes[0], axes[1], 100),
                         np.linspace(axes[2], axes[3], 100))
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=cmap)
    plt.contour(x1, x2, y_pred, cmap="Greys", alpha=0.8)
    colors = {"Wistia": ["#78785c", "#c47b27"], "Pastel1": ["red", "blue"]}
    markers = ("o", "^")
    for idx in (0, 1):
        plt.plot(X[:, 0][y == idx], X[:, 1][y == idx],
                 color=colors[cmap][idx], marker=markers[idx], linestyle="none")
    plt.axis(axes)
    plt.xlabel(r"$x_1$")
    plt.ylabel(r"$x_2$", rotation=0)

fig, axes = plt.subplots(ncols=2, figsize=(10, 4), sharey=True)
plt.sca(axes[0])
plot_decision_boundary(tree_clf1, X_moons, y_moons,
                       axes=[-1.5, 2.4, -1, 1.5], cmap="Wistia")
plt.title("Sense restriccions")
plt.sca(axes[1])
plot_decision_boundary(tree_clf2, X_moons, y_moons,
                       axes=[-1.5, 2.4, -1, 1.5], cmap="Wistia")
plt.title(f"min_samples_leaf = {tree_clf2.min_samples_leaf}")
plt.ylabel("")
plt.show()
```

```python
# Definim un nou conjunt de prova
X_moons_test, y_moons_test = make_moons(n_samples=1000, noise=0.2,
                                        random_state=43)
```

```python
# Veiem quina puntuació té el model sense regularitzar
tree_clf1.score(X_moons_test, y_moons_test)
```

```python
# I el regularitzat
tree_clf2.score(X_moons_test, y_moons_test)
```

# Regressió

```python
# Generació del conjunt de dades

from sklearn.tree import DecisionTreeRegressor

np.random.seed(42)
X_quad = np.random.rand(200, 1) - 0.5  # a single random input feature
y_quad = X_quad ** 2 + 0.025 * np.random.randn(200, 1)


# I del model 

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X_quad, y_quad)
```

```python
dibuixaArbre(tree_reg,feature_names=["x1"])
```

```python
# Definim altre model amb max_depth a 3:
tree_reg2 = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_reg2.fit(X_quad, y_quad)
```

```python
# I dibuixem els dos


def plot_regression_predictions(tree_reg, X, y, axes=[-0.5, 0.5, -0.05, 0.25]):
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    y_pred = tree_reg.predict(x1)
    plt.axis(axes)
    plt.xlabel("$x_1$")
    plt.plot(X, y, "b.")
    plt.plot(x1, y_pred, "r.-", linewidth=2, label=r"$\hat{y}$")

fig, axes = plt.subplots(ncols=2, figsize=(10, 4), sharey=True)
plt.sca(axes[0])
plot_regression_predictions(tree_reg, X_quad, y_quad)

th0, th1a, th1b = tree_reg.tree_.threshold[[0, 1, 4]]
for split, style in ((th0, "k-"), (th1a, "k--"), (th1b, "k--")):
    plt.plot([split, split], [-0.05, 0.25], style, linewidth=2)
plt.text(th0, 0.16, "Depth=0", fontsize=15)
plt.text(th1a + 0.01, -0.01, "Depth=1", horizontalalignment="center", fontsize=13)
plt.text(th1b + 0.01, -0.01, "Depth=1", fontsize=13)
plt.ylabel("$y$", rotation=0)
plt.legend(loc="upper center", fontsize=16)
plt.title("max_depth=2")

plt.sca(axes[1])
th2s = tree_reg2.tree_.threshold[[2, 5, 9, 12]]
plot_regression_predictions(tree_reg2, X_quad, y_quad)
for split, style in ((th0, "k-"), (th1a, "k--"), (th1b, "k--")):
    plt.plot([split, split], [-0.05, 0.25], style, linewidth=2)
for split in th2s:
    plt.plot([split, split], [-0.05, 0.25], "k:", linewidth=1)
plt.text(th2s[2] + 0.01, 0.15, "Depth=2", fontsize=13)
plt.title("max_depth=3")

plt.show()

```

```python
# Prediccions amb un arbre sense regularitzar i un regularitzat

tree_reg1 = DecisionTreeRegressor(random_state=42)
tree_reg2 = DecisionTreeRegressor(random_state=42, min_samples_leaf=10)
tree_reg1.fit(X_quad, y_quad)
tree_reg2.fit(X_quad, y_quad)

x1 = np.linspace(-0.5, 0.5, 500).reshape(-1, 1)
y_pred1 = tree_reg1.predict(x1)
y_pred2 = tree_reg2.predict(x1)

fig, axes = plt.subplots(ncols=2, figsize=(10, 4), sharey=True)

plt.sca(axes[0])
plt.plot(X_quad, y_quad, "b.")
plt.plot(x1, y_pred1, "r.-", linewidth=2, label=r"$\hat{y}$")
plt.axis([-0.5, 0.5, -0.05, 0.25])
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.legend(loc="upper center")
plt.title("No restrictions")

plt.sca(axes[1])
plt.plot(X_quad, y_quad, "b.")
plt.plot(x1, y_pred2, "r.-", linewidth=2, label=r"$\hat{y}$")
plt.axis([-0.5, 0.5, -0.05, 0.25])
plt.xlabel("$x_1$")
plt.title(f"min_samples_leaf={tree_reg2.min_samples_leaf}")

plt.show()
```

```python
# Creem un Random Forest
from sklearn.ensemble import RandomForestRegressor

random_reg = RandomForestRegressor()
random_reg.fit(X_quad, y_quad)
```

```python
def plot_rf_predictions(rf_reg, X, y, axes=[-0.5, 0.5, -0.05, 0.25]):
    x1 = np.linspace(axes[0], axes[1], 500).reshape(-1, 1)
    y_pred = rf_reg.predict(x1)
    plt.axis(axes)
    plt.xlabel("$x_1$")
    plt.plot(X, y, "b.")
    plt.plot(x1, y_pred, "r.-", linewidth=2, label=r"$\hat{y}$")

plot_rf_predictions(random_reg, X_quad, y_quad)
plt.ylabel("$y$", rotation=0)
plt.legend(loc="upper center", fontsize=16)
plt.title("Random Forest")
plt.show()
```
